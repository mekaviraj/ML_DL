```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from transformers import AutoTokenizer, TFAutoModel
from google import genai
from google.genai import types
```

```python
#3

#  Dataset Simulation and Strategy

np.random.seed(42)

CLASSES = ["No Finding", "Pneumonia", "Effusion", "Atelectasis", "Cardiomegaly"]

num_samples = 200
mock_data = {
    "Image Index": [f"images_{i:03d}.png" for i in range(num_samples)],
    "Finding Labels": np.random.choice(CLASSES, size=num_samples, p=[0.4, 0.15, 0.15, 0.15, 0.15])}

df = pd.DataFrame(mock_data)

os.makedirs("mock_images", exist_ok=True)
for img_name in df["Image Index"]:
    dummy_img = np.random.randint(0, 255, (224, 224, 3), dtype=np.uint8)
    cv2.imwrite(f"mock_images/{img_name}", dummy_img)

print(f"Dataset simulated successfully. Shape: {df.shape}")
print(df["Finding Labels"].value_counts())


# Define our 5 target classes
# Generate a mock dataframe mirroring the NIH layout
# Create a mock image directory and generate dummy X-ray tensors
# Simulating a grayscale chest X-ray image array


# mimic the nih cuz free colab cannot handle 100gb dataprocessing
```

```python
# 4: Synthetic Text Generation
symptom_map = { "No Finding": "Patient presents for routine screening. Clear lung fields. No acute distress reported.",
    "Pneumonia": "Patient reports high fever, productive cough with thick sputum, and localized chest pain. Audible crackles during auscultation.",
    "Effusion": "Patient complains of progressive shortness of breath and dull chest pain worse on deep inspiration. Decreased breath sounds at bases.",
    "Atelectasis": "Post-operative patient exhibiting shallow breathing and minor hypoxemia. Diminished breath sounds in lower lobes.",
    "Cardiomegaly": "Patient reports chronic fatigue, peripheral ankle edema, and orthopnea. Notable cardiothoracic silhouette expansion noted historically."}

df["Clinical_Notes"] = df["Finding Labels"].map(symptom_map)

labels_encoded = pd.get_dummies(df["Finding Labels"])[CLASSES].values

print("Sample Multi-Modal Data Row:")

print(f"Label: {df['Finding Labels'].iloc[1]}\nNote: {df['Clinical_Notes'].iloc[1]}")

# One-hot encode labels for model training
# Apply mapping to create text features
```

```python
# 5: Input Tokenization and Feature Prep
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

# Tokenize text descriptions
tokenized_text = tokenizer(
    list(df["Clinical_Notes"]),
    padding="max_length",
    truncation=True,
    max_length=32,
    return_tensors="np"
)

input_ids = tokenized_text["input_ids"]
attention_masks = tokenized_text["attention_mask"]

# Load and preprocess images into an array
image_data = []
for img_name in df["Image Index"]:
    img_path = f"mock_images/{img_name}"
    img = cv2.imread(img_path)
    img = cv2.resize(img, (224, 224))
    image_data.append(img)

image_data = np.array(image_data, dtype=np.float32) / 255.0
print(f"Images tensor shape: {image_data.shape}")
print(f"Text Input IDs shape: {input_ids.shape}")
```

```python
# 6: Network Architecture Setup
def build_multimodal_model():
    # 1. Image Branch
    image_input = layers.Input(shape=(224, 224, 3), name="image_input")
    base_image_model = tf.keras.applications.EfficientNetB0(
        include_top=False, weights=None, pooling="avg"
    )(image_input)
    image_dense = layers.Dense(256, activation="relu")(base_image_model)

    # 2. Text Branch (DistilBERT)
    text_input_ids = layers.Input(shape=(32,), dtype=tf.int32, name="text_input_ids")
    text_attention_mask = layers.Input(shape=(32,), dtype=tf.int32, name="text_attention_mask")

    transformer_model = TFAutoModel.from_pretrained("distilbert-base-uncased")
    # Freeze DistilBERT to conserve Free Colab memory resources during initial phases
    transformer_model.trainable = False

    transformer_outputs = transformer_model(input_ids=text_input_ids, attention_mask=text_attention_mask)
    text_features = transformer_outputs[0][:, 0, :] # Extract CLS token
    text_dense = layers.Dense(256, activation="relu")(text_features)

    # 3. Multimodal Fusion Layer
    fused_features = layers.Concatenate()([image_dense, text_dense])

    # 4. Classification Output Head
    x = layers.Dense(128, activation="relu")(fused_features)
    x = layers.Dropout(0.3)(x)
    output = layers.Dense(5, activation="softmax", name="disease_output")(x)

    model = models.Model(inputs=[image_input, text_input_ids, text_attention_mask], outputs=output)
    return model

multimodal_sys = build_multimodal_model()
multimodal_sys.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
multimodal_sys.summary()
```

```python
#7: Execution of Model Training
history = multimodal_sys.fit(
    x={
        "image_input": image_data,
        "text_input_ids": input_ids,
        "text_attention_mask": attention_masks
    },
    y=labels_encoded,
    epochs=10, # Kept low for immediate execution testing
    batch_size=16,
    validation_split=0.1
)
```

```python
# 8: Grad-CAM Visualization Setup — Full Eager Execution Approach
def generate_gradcam(model, img_array, text_id, text_mask, intensity=0.5, res=224):
    """
    Grad-CAM using pure eager execution.
    Avoids all symbolic graph wiring issues caused by EfficientNetB0
    being called as a black-box layer (pooling='avg') inside the multimodal model.
    Strategy: watch a tf.Variable image tensor, run the full model forward,
    then backprop to the image. Extract spatial importance via the gradient map.
    """
    # Cast inputs to the right types
    img_tensor  = tf.cast(img_array, tf.float32)       # (1, 224, 224, 3)
    text_id     = tf.cast(text_id,   tf.int32)
    text_mask   = tf.cast(text_mask, tf.int32)

    # Wrap image as a Variable so GradientTape can differentiate w.r.t. it
    img_var = tf.Variable(img_tensor, trainable=True)

    with tf.GradientTape() as tape:
        tape.watch(img_var)
        predictions = model([img_var, text_id, text_mask], training=False)
        class_idx   = tf.argmax(predictions[0])
        loss        = predictions[:, class_idx]

    # Gradient of the winning class score w.r.t. every input pixel
    grads = tape.gradient(loss, img_var)               # (1, 224, 224, 3)

    if grads is None:
        raise ValueError("Gradients are None — tape.watch() on img_var failed.")

    # Collapse channel dim → spatial saliency map  (224, 224)
    grad_cam = tf.reduce_mean(tf.abs(grads[0]), axis=-1)  # (224, 224)

    # ReLU + safe normalisation
    grad_cam = tf.maximum(grad_cam, 0)
    max_val  = tf.reduce_max(grad_cam)
    grad_cam = grad_cam / (max_val + tf.keras.backend.epsilon())
    heatmap  = grad_cam.numpy()                        # (224, 224) float32 0-1

    # ── Overlay ──────────────────────────────────────────────────────────────
    img             = (img_array[0] * 255.0).astype(np.float32)   # (224,224,3)
    heatmap_resized = cv2.resize(heatmap, (res, res))
    heatmap_uint8   = np.uint8(255 * heatmap_resized)
    heatmap_colored = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)

    overlay = heatmap_colored.astype(np.float32) * intensity + img
    overlay = np.clip(overlay, 0, 255).astype(np.uint8)

    return heatmap_colored, overlay


# ── Smoke-test on sample 0 ────────────────────────────────────────────────────
test_img  = np.expand_dims(image_data[0],     axis=0)
test_id   = np.expand_dims(input_ids[0],      axis=0)
test_mask = np.expand_dims(attention_masks[0], axis=0)

heatmap, output_image = generate_gradcam(
    multimodal_sys, test_img, test_id, test_mask
)

print("✅ Grad-CAM computations completed successfully.")
print(f"   heatmap shape : {heatmap.shape}")
print(f"   overlay shape : {output_image.shape}")
```

```python
# 9: Clinical Explanation Generator
from google.colab import userdata

USE_GEMINI = True  # Set to True to fetch from Colab Secrets

# High-fidelity baseline local medical templates
local_templates = {
    "No Finding": "The predictive model indicates normal baseline physiology. No acute clinical anomalies or structural damage observed in lung fields.",
    "Pneumonia": "The system isolated standard signs consistent with consolidation patterns or acute infectious fluid accumulation in alveolar channels.",
    "Effusion": "The network highlighted markers indicative of fluid pooling within the pleural cavity, narrowing efficient respiratory paths.",
    "Atelectasis": "Data parameters line up with partial or full collapsing zones within localized pulmonary lobules.",
    "Cardiomegaly": "Visual indicators trace a clear enlargement pattern of the cardiac profile exceeding normal diagnostic thresholds."
}

def generate_medical_explanation(disease, confidence):
    prompt = f"The AI analysis engine predicted {disease} with {confidence*100:.1f}% tracking confidence. Give a brief, 2-sentence medical explanation of this finding."

    if USE_GEMINI:
        try:
            # Safely retrieve your key directly from Colab Secrets
            gemini_key = userdata.get('GEMINI_API_KEY')

            client = genai.Client(api_key=gemini_key)
            response = client.models.generate_content(
                model='gemini-2.5-flash',
                contents=prompt,
            )
            return response.text.strip()

        except Exception as e:
            # Drop back smoothly if the secret doesn't exist
            return f"[API Fallback Mode]: {local_templates[disease]}"
    else:
        return f"[Fallback Template Mode]: {local_templates[disease]}"

print("Explanation router configured with Colab Secrets integration.")
```

```python
#10: Model Evaluation Metrics
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Compute prediction matrices across the entire validation split/subset
val_preds = multimodal_sys.predict([image_data, input_ids, attention_masks])
val_pred_classes = np.argmax(val_preds, axis=1)
true_classes = np.argmax(labels_encoded, axis=1)

# Generate Classification Matrix Report
print("\n" + "="*60)
print("             MULTIMODAL SYSTEM PERFORMANCE REPORT")
print("="*60)
print(classification_report(true_classes, val_pred_classes, target_names=CLASSES))

# Generate and plot Confusion Matrix
cm = confusion_matrix(true_classes, val_pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=CLASSES, yticklabels=CLASSES)
plt.title('Disease Classification Confusion Matrix Target Analysis')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.show()
```
