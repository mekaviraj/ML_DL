```python
import tensorflow as tf
from tensorflow.keras import layers, models
from transformers import AutoTokenizer, TFAutoModel

# Custom Keras Layer to wrap the TFAutoModel
class TextEmbeddingLayer(layers.Layer):
    def __init__(self, model_name, **kwargs):
        super().__init__(**kwargs)
        self.transformer_model = TFAutoModel.from_pretrained(model_name)
        self.transformer_model.trainable = False  # Freeze the model

    def call(self, inputs):
        input_ids, attention_mask = inputs
        # Ensure return_dict=False to get a tuple output that Keras can handle
        transformer_outputs = self.transformer_model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            return_dict=False
        )
        return transformer_outputs[0][:, 0, :]  # Extract CLS token

def build_multimodal_model():
    # --- Image branch: same weak CNN as baseline ---
    image_input = layers.Input(shape=(224, 224, 3), name="image_input")
    x = layers.Conv2D(16, 3, activation="relu", padding="same")(image_input)
    x = layers.MaxPooling2D(4)(x)
    x = layers.Conv2D(32, 3, activation="relu", padding="same")(x)
    x = layers.MaxPooling2D(4)(x)
    x = layers.Conv2D(32, 3, activation="relu", padding="same")(x)
    x = layers.GlobalAveragePooling2D()(x)
    image_features = layers.Dense(128, activation="relu")(x)

    # --- Text branch: DistilBERT (unchanged) ---
    text_input_ids     = layers.Input(shape=(32,), dtype=tf.int32, name="text_input_ids")
    text_attention_mask = layers.Input(shape=(32,), dtype=tf.int32, name="text_attention_mask")
    transformer_model  = TFAutoModel.from_pretrained("distilbert-base-uncased")
    transformer_model.trainable = False
    transformer_outputs = transformer_model(
        input_ids=text_input_ids, attention_mask=text_attention_mask
    )
    text_features = transformer_outputs[0][:, 0, :]
    text_features = layers.Dense(128, activation="relu")(text_features)

    # --- Fusion ---
    fused = layers.Concatenate()([image_features, text_features])
    x = layers.Dense(128, activation="relu")(fused)
    x = layers.Dropout(0.3)(x)
    output = layers.Dense(5, activation="softmax", name="disease_output")(x)

    model = models.Model(
        inputs=[image_input, text_input_ids, text_attention_mask],
        outputs=output
    )
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-4),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model
# def build_multimodal_model():
#     # 1. Image Branch
#     image_input = layers.Input(shape=(224, 224, 3), name="image_input")
#     base_image_model = tf.keras.applications.EfficientNetB0(
#         include_top=False, weights=None, pooling="avg"
#     )(image_input)
#     image_dense = layers.Dense(256, activation="relu")(base_image_model)

#     # 2. Text Branch (DistilBERT)
#     text_input_ids = layers.Input(shape=(32,), dtype=tf.int32, name="text_input_ids")
#     text_attention_mask = layers.Input(shape=(32,), dtype=tf.int32, name="text_attention_mask")

#     # Use the custom layer to wrap the transformer model
#     text_features = TextEmbeddingLayer("distilbert-base-uncased")(
#         [text_input_ids, text_attention_mask]
#     )
#     text_dense = layers.Dense(256, activation="relu")(text_features)

#     # 3. Multimodal Fusion Layer
#     fused_features = layers.Concatenate()([image_dense, text_dense])

#     # 4. Classification Output Head
#     x = layers.Dense(128, activation="relu")(fused_features)
#     x = layers.Dropout(0.3)(x)
#     output = layers.Dense(5, activation="softmax", name="disease_output")(x)

#     model = models.Model(inputs=[image_input, text_input_ids, text_attention_mask], outputs=output)
#     return model

multimodal_sys = build_multimodal_model()
multimodal_sys.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
multimodal_sys.summary()
```

```python
# 8: Grad-CAM Visualization Setup — Full Eager Execution Approach
def generate_gradcam(model, img_array, text_id, text_mask, intensity=0.5, res=224):
    """
    Grad-CAM using pure eager execution.
    Avoids all symbolic graph wiring issues caused by EfficientNetB0
    being called as a black-box layer (pooling='avg') inside the multimodal model.
    Strategy: watch a tf.Variable image tensor, run the full model forward,
    then backprop to the image. Extract spatial importance via the gradient map.
    """
    # Cast inputs to the right types
    img_tensor  = tf.cast(img_array, tf.float32)       # (1, 224, 224, 3)
    text_id     = tf.cast(text_id,   tf.int32)
    text_mask   = tf.cast(text_mask, tf.int32)

    # Wrap image as a Variable so GradientTape can differentiate w.r.t. it
    img_var = tf.Variable(img_tensor, trainable=True)

    with tf.GradientTape() as tape:
        tape.watch(img_var)
        predictions = model([img_var, text_id, text_mask], training=False)
        class_idx   = tf.argmax(predictions[0])
        loss        = predictions[:, class_idx]

    # Gradient of the winning class score w.r.t. every input pixel
    grads = tape.gradient(loss, img_var)               # (1, 224, 224, 3)

    if grads is None:
        raise ValueError("Gradients are None — tape.watch() on img_var failed.")

    # Collapse channel dim → spatial saliency map  (224, 224)
    grad_cam = tf.reduce_mean(tf.abs(grads[0]), axis=-1)  # (224, 224)

    # ReLU + safe normalisation
    grad_cam = tf.maximum(grad_cam, 0)
    max_val  = tf.reduce_max(grad_cam)
    grad_cam = grad_cam / (max_val + tf.keras.backend.epsilon())
    heatmap  = grad_cam.numpy()                        # (224, 224) float32 0-1

    # ── Overlay ──────────────────────────────────────────────────────────────
    img             = (img_array[0] * 255.0).astype(np.float32)   # (224,224,3)
    heatmap_resized = cv2.resize(heatmap, (res, res))
    heatmap_uint8   = np.uint8(255 * heatmap_resized)
    heatmap_colored = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)

    overlay = heatmap_colored.astype(np.float32) * intensity + img
    overlay = np.clip(overlay, 0, 255).astype(np.uint8)

    return heatmap_colored, overlay


# ── Smoke-test on sample 0 ────────────────────────────────────────────────────
test_img  = np.expand_dims(image_data[0],     axis=0)
test_id   = np.expand_dims(input_ids[0],      axis=0)
test_mask = np.expand_dims(attention_masks[0], axis=0)

heatmap, output_image = generate_gradcam(
    multimodal_sys, test_img, test_id, test_mask
)

print("✅ Grad-CAM computations completed successfully.")
print(f"   heatmap shape : {heatmap.shape}")
print(f"   overlay shape : {output_image.shape}")
```
